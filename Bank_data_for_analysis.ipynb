```python
bank_data['duration'] = bank_data['duration']/60
duration_list = pd.cut(bank_data['duration'],[0,0.1,0.2,0.3,0.4,100]).rename('dur_interval')
duration_list = pd.concat([duration_list,bank_data['y']], axis = 1)         

sns.factorplot(data = duration_list, x='dur_interval',hue ='y', kind = 'count')
plt.xlabel('Duration of call (in mins)')
plt.ylabel('No of Calls made')
plt.title('Calls effectiveness w.r.t. duration')
plt.subplots_adjust(top = 0.9)
```

## Categorizing the data frame
Categorizing the dataframe and convert it numeric 
Label Encoding the data frame

```python
label_data = final_data
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
label_data['job'] = le.fit_transform(label_data['job'])
label_data['marital'] = le.fit_transform(label_data['marital'])
# label_data['default'] = le.fit_transform(label_data['default'])
label_data['housing'] = le.fit_transform(label_data['housing'])
label_data['month'] = le.fit_transform(label_data['month'])
label_data['loan'] = le.fit_transform(label_data['loan'])
# label_data['contact'] = le.fit_transform(label_data['contact'])
label_data['day_of_week'] = le.fit_transform(label_data['day_of_week'])
# label_data['poutcome'] = le.fit_transform(label_data['poutcome'])
label_data['y'] = le.fit_transform(label_data['y'])
```
